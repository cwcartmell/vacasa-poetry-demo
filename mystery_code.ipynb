```python
import random
import altair
import numpy
import pandas
import math
```

```python
def jitter(x):
    return x+random.random()


def line_df(a, b):
    
    dist = int(math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2))
    density = 25
    
    X = numpy.linspace(a[0], b[0], dist*density)
    Y = numpy.linspace(a[1], b[1], dist*density)
    
    X = numpy.array([x+random.random() for x in X])
    Y = numpy.array([y+random.random() for y in Y])
    
    return pandas.DataFrame({"x": X, "y": Y})


def connected_lines_df(points):
    return pandas.concat(
        [line_df(points[i], points[i+1]) for i in range(0, len(points)-1)],
    )


outer_points = ((4, 0), (0,0), (0,6), (5, 10), (10, 6), (10, 0), (6, 0))
middle_points = ((2, 2), (2, 5), (5, 7), (8, 5), (8, 2))
inner_points = ((4, 2), (4, 4), (5, 5), (6, 4), (6, 2))

points = pandas.concat([
    connected_lines_df(outer_points),
    connected_lines_df(middle_points),
    connected_lines_df(inner_points),
])

altair.Chart(points, mark="point").encode(
    x='x:Q',
    y='y:Q',
)
```
